```python
# permutations using library function 
from itertools import permutations 
import numpy as np
  
# Get all permutations of [1, 2, 3] 
perm = permutations([i for i in range(8)])

def testPermuation(p):
    if not (p.index(6) < p.index(7)):
        return False
    
    if not (p.index(2) < p.index(3) < p.index(5)):
        return False
    
    if not (p.index(1) > p.index(3)):
        return False
    
    
    if not (p[0] == 6):
        return False
    
    if not (p.index(4) < p.index(2)):
        return False
    if not (p.index(4) < p.index(0)):
        return False
    
    if not (p.index(6) < p.index(0)):
        return False
    if not (p.index(7) < p.index(1)):
        return False
    
    if (p.index(0) > p.index(2)) and (p.index(1) < p.index(5)):
        return False
    
    if not (p.index(7) < p.index(5)):
        return False
    
    # strange restriction
    if (p.index(2) < p.index(7)) and (p.index(2) < p.index(0)):
        return False
    
    return True

ret12 = []
for p in perm:
    if testPermuation(p):
        ret12.append(p)
```

```python
def takeIntersection(s):
    ret = []
    for i in range(len(s[0])):
        t = [set(x[i]) for x in s]
        r = set()
        for tt in t:
            r |= set(tt)
        r = tuple(r)
        ret.append(r)
        
    return ret
```

```python
def isContained(sub,sup):
    for i in range(len(sub)):
        if set(sub[i]).issuperset(set(sup[i])):
            continue
        else:
            return False
    return True

#dl = [((6,),(0,),(0,),()), ((),(6,),(),()), ((0,),(),(),()),((),(),(),(0,))]
#tl = [((7,),(),(),(1,)),((),(7,),(),()),((),(),(1,),())]
#rl = [((),(2,),(2,),(3,)), ((2,),(),(),()), ((),(),(),(2,)),((),(),(3,),())]
#s1l = [((4,),(),(),()), ((),(4,),(),())]
#s2l = [((),(4,),(4,),()), ((4,),(),(),()), ((),(),(),(4,))]
#s3l = [((),(),(),(5,)), ((),(),(5,),())]

def identifyIndex8(l):
    d8 = ((6,),(0,),(0,),())
    t8 = ((7,),(),(),(1,))
    r8 = ((),(2,),(2,),(3,))
    s18 = ((4,),(),(),())
    s28 = ((),(4,),(4,),())
    s38 = ((),(),(),(5,))
    
    indexList = [d8,t8,r8,s18,s28,s38]
    portrait=[]
    for region in indexList:
        if isContained(l,region):
            portrait.append(1)
        else:
            portrait.append(0)           
    return tuple(portrait)
```

```python
partialOrderBase =[(6,7),(2,3),(3,5),(3,1),(6,1),(6,2),(6,3),(6,4),(6,5),(6,0),(4,2),(4,0),(7,1),(7,5)] 
prec = dict()
desc = dict()
for i,j in partialOrderBase:
    if i not in desc:
        desc[i] = []
    desc[i].append(j)
    
    if j not in prec:
        prec[j] = []
    prec[j].append(i)
    
def genPrecDict(i,prec):
    
    if i not in prec:
        return set()
    
    ret = set(prec[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in prec:
                ret |= set(prec[j])
        
        if ret == preRet:
            return ret
        
def genDescDict(i,desc):
    
    if i not in desc:
        return set()
    
    ret = set(desc[i])
    while True:
        preRet = ret.copy()
        for j in preRet:
            if j in desc:
                ret |= set(desc[j])
        
        if ret == preRet:
            return ret

preDict = dict()
desDict = dict()

for i in range(8):
    preDict[i] = genPrecDict(i,prec)
    desDict[i] = genDescDict(i,desc)
```

```python
# simplify the defining inequalities
def getMaixmal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in desDict:
            intersect = set(l)&set(desDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret

def getMinimal(l,desDict):
    ret = []
    for i in l:
        intersect = set()
        if i in preDict:
            intersect = set(l)&set(preDict[i])
        if len(intersect) ==0:
            ret.append(i)    
    return ret
```

```python
def removeRedundant(l):
    ret = []
    for i in range(len(l)):
        flag = True
        for j in range(len(l)):
            if i!=j and isContained(l[i],l[j]):
                flag = False
                break
        if flag:
            ret.append(l[i])
    return ret
```

```python
def generatePairConnectedComponent(indexSet,atomicComponents):
    fUnion = []
    for i in range(len(indexSet)):
        curr = indexSet[i]
        toRemove = []
        for j in range(len(fUnion)):
            currUnion = fUnion[j]
            for k in range(len(currUnion)):
                if isConnected(indexSet[i],currUnion[k],atomicComponents):
                    #print(indexSet[i],currUnion[k])
                    toRemove.append(j)
                    break
        newUnion = [curr]
        for j in range(len(toRemove)):
            index = toRemove[j]
            newUnion+=fUnion[index]
        updatedUnion = []
        for j in range(len(fUnion)):
            if j not in toRemove:
                updatedUnion.append(fUnion[j])
        updatedUnion.append(newUnion)
        fUnion = updatedUnion
           
    return fUnion

def isOrderContained(order, po):
    for i in po[0]:
        if order.index(i) > order.index(8):
            return False
    
    for i in po[1]:
        if order.index(i) < order.index(8):
            return False
    
    for i in po[2]:
        if order.index(i) > order.index(9):
            return False
        
    for i in po[3]:
        if order.index(i) < order.index(9):
            return False
        
    return True

def isConnected(c1,c2, atomicComponents):
    intersection = takeIntersection([c1,c2])
    for atom in atomicComponents:
        if isOrderContained(atom,intersection):
            #print(atom)
            #print(intersection)
            return True
    return False
```

```python
def isAdjacentOrder(order1, order2):
    differ = []
    for i in range(len(order1)):
        if order1[i]!= order2[i]:
            differ.append(i)
            
    if len(differ)!= 2:
        return False
    
    if abs(differ[0] - differ[1]) == 1:
        return True
    else:
        return False

def isAdjacentOrders(orders1, orders2):
    for order1 in orders1:
        for order2 in orders2:
            if isAdjacentOrder(order1, order2):
                return True
    return False
```
